#  Análisis de Componentes Principales (PCA)    

```r
x_scaled <- scale(x[,-1])
x_scaled
```

```r




pca_model <- prcomp(x_scaled, scale. = TRUE, center= TRUE)
summary(pca_model)

```

```r
biplot(pca_model, scale = 0)
```

```r
var_exp <- pca_model$sdev^2 / sum(pca_model$sdev^2)
var_exp
```

```r
library(ggplot2)
qplot(1:length(var_exp), var_exp, geom = "line") + 
  labs(x = "Componentes Principales", y = "Varianza Explicada")
```

```r
plot(pca_model)
```

```r
library(factoextra)
fviz_pca_var(pca_model, col.var = "blue", axes = c(1, 2), repel = TRUE)
```

```r
explained_variance <- summary(pca_model)$importance[2, 1:3] * 100
explained_variance
```

```r
loadings <- pca_model$rotation[,1:3]
loadings
```

### Interpretación de los Componentes Principales

- **PC1**: Captura una dimensión relacionada con **confianza**, **beneficios** y aspectos de **servicio personalizado** y **productos financieros**. Es un componente que refleja una **actitud general positiva** hacia la relación con la empresa o el servicio.

- **PC2**: Está claramente asociado con la **edad** de los clientes, mientras que otras características, como **confianza**, **beneficios** y **productos**, tienen una relación mínima con este componente.

- **PC3**: Está relacionado con características de **contacto con ejecutivos**, **relevancia de los productos** y **seguridad**. Además, parece haber una **relación inversa** con las variables de **atención personalizada** y **productos personalizados**, lo que sugiere que los clientes que perciben estas características pueden estar más enfocados en otros aspectos, como el **contacto directo con los ejecutivos** o **la seguridad**.


```r
scores <- pca_model$x[,1:3]
scores
```

# Clustering

```r
library(factoextra) 
fviz_nbclust(scores, kmeans, method = "wss") + 
  ggtitle("Método de Codo para Determinar el Número de Clusters")
```

```r
#Aplicar K-means con el número óptimo de clusters
set.seed(123)
clusters <- kmeans(scores, centers = 3)
clusters
```

```r
clusters$cluster
clusters$centers
```

```r
Segmentos <- clusters$cluster %>% as.data.frame()
colnames(Segmentos) <- c("SEGMENTO")
```

```r
Data_final <- cbind(x,scores,Segmentos)
Data_final
```

```r


# Crear gráfico de dispersión
ggplot(Data_final, aes(x = PC1, y = PC2, color = SEGMENTO)) +
  geom_point(size = 3) +
  labs(title = "Clustering K-means en PCA", x = "PC1", y = "PC2") +
  theme_minimal()
```

```r
resumen_clusters <- Data_final %>%
  group_by(SEGMENTO) %>%
  summarise(across(everything(), list(media = ~mean(.x, na.rm = TRUE))))

resumen_clusters
```

### Interpretación de los Clusters

#### **Cluster 1: Clientes con alta satisfacción**
- **Descripción**: Este grupo de clientes tiene una **alta percepción positiva** sobre la calidad de servicio y productos ofrecidos.
- **Características destacadas**:
  - Alta **confianza** en la empresa.
  - Valoran la **rapidez** de las operaciones, la **personalización** de productos y la **asesoría experta**.
  - Están muy enfocados en productos de **crédito**, **seguridad** y **beneficios**.
  - Aprecian la **accesibilidad de la información** y el **contacto directo con los ejecutivos**.
- **Conclusión**: Son clientes comprometidos y satisfechos, con una visión positiva hacia los servicios ofrecidos.

#### **Cluster 2: Clientes con baja satisfacción**
- **Descripción**: Este grupo está compuesto por clientes con **bajas expectativas** sobre la calidad de servicio, los beneficios y los productos.
- **Características destacadas**:
  - Menor satisfacción con la **rapidez** de las operaciones y la **personalización** de los productos.
  - Perciben que la **atención personalizada** y los productos no cumplen sus expectativas.
  - Este grupo probablemente tiene un **menor compromiso** y **satisfacción** con la empresa.
- **Conclusión**: Son clientes que necesitan mejoras en la calidad de servicio y la personalización para aumentar su satisfacción y compromiso.

#### **Cluster 3: Clientes con expectativas moderadas**
- **Descripción**: Este grupo muestra una **percepción moderada** de los servicios y productos, con una valoración equilibrada.
- **Características destacadas**:
  - Tienen un enfoque equilibrado en aspectos como la **confianza**, **beneficios**, **rapidez**, **productos de crédito** y **asesoría experta**.
  - Aunque valoran estos aspectos, su percepción es **más neutra** en comparación con el Cluster 1.
- **Conclusión**: Son clientes relativamente satisfechos con la empresa, pero con **expectativas más moderadas**. Este grupo podría beneficiarse de un enfoque para incrementar su satisfacción y compromiso.

---

### Resumen General:
- **Cluster 1**: Clientes altamente satisfechos y comprometidos, con una percepción muy positiva.
- **Cluster 2**: Clientes insatisfechos y con bajas expectativas, que podrían necesitar atención mejorada.
- **Cluster 3**: Clientes con expectativas moderadas, relativamente satisfechos, pero con espacio para mejorar su percepción.
